#Part2

```python
from openai.embeddings_utils import get_embedding, cosine_similarity
```

```python
df = pd.read_csv('new_words.csv')
print(df)
```

```python
len(df)
```

```python
df['embedding'] = df['text'].apply(lambda x: get_embedding(x, engine='text-embedding-ada-002'))
df.to_csv('word-em.csv')
```

```python
DF = pd.read_csv('word-em.csv')
DF['embedding'] = DF['embedding'].apply(eval).apply(np.array)
DF
```

```python
type(DF)
```

```python
search = input('please input anything you wanna search!')
```

```python
search2vec = get_embedding(search, engine="text-embedding-ada-002")
search2vec
```

```python
DF["similarities"] = DF['embedding'].apply(lambda x: cosine_similarity(x, search2vec))
DF.sort_values("similarities", ascending=False)
```

```python
text1 = "I like you"
text2 = "I love you。"
text3 = "I dislike you。"
emb1 = get_embedding(text1, "text-embedding-ada-002")
emb2 = get_embedding(text2, "text-embedding-ada-002")
emb3 = get_embedding(text3, "text-embedding-ada-002")
```

```python
cosine_similarity(emb1, emb2)
```

```python
cosine_similarity(emb1, emb3)
```

```python
cosine_similarity(emb2, emb3)
```

```python
content = "please show me the similarity among these sentences：\n1. I like you。\n2.I love you。\n3.get out, you are nothing to me!\n"
```

```python
completion = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": content}]
)
print(completion.choices[0].message)
```
